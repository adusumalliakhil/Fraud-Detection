```r
# Importing the dataset
dataset <- read.csv("D:\\projectR\\datafiles\\creditcard.csv", header = TRUE,
                stringsAsFactors = FALSE,
                strip.white = TRUE ,
                sep = ',')
```

```r
dataset$Class = as.factor(dataset$Class)
```

```r
#install.packages("ROSE")
#Implementing undersampling
library(ROSE)
dataset = ovun.sample(Class~., data = dataset, method = 'over', 
                       N =2*284315 , seed = 1234)$data
summary(dataset$Class)
```

```r
dataset$Class = as.numeric(dataset$Class)
```

```r
t.test(dataset)
```

```r
library(corrplot)
co<-cor(dataset)
corrplot(co, method="pie")
```
