```r
# Importing the dataset
dataset <- read.csv("D:\\projectR\\datafiles\\creditcard.csv", header = TRUE,
                stringsAsFactors = FALSE,
                strip.white = TRUE ,
                sep = ',')
```

```r
summary(dataset)
```

```r
library(psych)
data2 <- describe(dataset)
data2 <- data2[c(4,10,11,12,13)]
data2
```

```r
library(purrr)
library(tidyr)
library(ggplot2)
par(mfrow=c(16,2))
dataset %>%
  keep(is.numeric) %>% 
  gather() %>% 
  ggplot(aes(value)) +
    facet_wrap(~ key, scales = "free") +
    geom_histogram()
```

```r
library(ggplot2)
dataset %>%
  keep(is.numeric) %>%                     # Keep only numeric columns
  gather() %>%                             # Convert to key-value pairs
  ggplot(aes(value)) +                     # Plot the values
    facet_wrap(~ key, scales = "free") +   # In separate panels
    geom_density()                         # as density
```

```r
data3 <- dataset[c(-1,-30)]
boxplot(data3,
        main = toupper("Boxplot of creditcard"),
        ylab = "Frequency",
        col = "blue")
```
